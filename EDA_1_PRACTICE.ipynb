```python
# Exploratory Data Analysis

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/slmsshk/DataSet/main/data_clean.csv",index_col=0)
```

```python
# data type conversion

data.info()
```

```python
data['Month']=pd.to_numeric(data['Month'],errors='coerce')
data['Temp C']=pd.to_numeric(data['Temp C'],errors='coerce')
data['Weather']=data['Weather'].astype('category')
```

```python
data.info()

```

```python
#Duplicates

data.duplicated()
```

```python
# count the duplicated rows
data[data.duplicated()].shape
```

```python
# print the duplicated rows
data[data.duplicated()]
```

```python
data_cleaned1=data.drop_duplicates()
```

```python
data_cleaned1.shape
```

```python
# drop columns
data
```

```python
data_cleaned2=data_cleaned1.drop('Temp C',axis=1)
```

```python
data_cleaned2
```

```python
# rename the solar column
data_cleaned3=data_cleaned2.rename({'Solar.R':'Solar'},axis=1)
```

```python
data_cleaned3
```

```python

# missing values and imputations

import seaborn as sns
```

```python
cols=data_cleaned3.columns
colours=['#000099','#ffff00']
sns.heatmap(data_cleaned3[cols].isnull(),cmap=sns.color_palette(colours))
```

```python
data_cleaned3[data_cleaned3.isnull().any(axis=1)].head()
```

```python
data_cleaned3.isnull().sum()
```

```python
# mean tmputation

mean=data_cleaned3['Ozone'].mean()
print(mean)
```

```python
data_cleaned3['Ozone']=data_cleaned3['Ozone'].fillna(mean)
```

```python
# Box plot
data_box=data_cleaned3.dropna()
data1_box=data_box.Ozone
plt.boxplot(data1_box)
```
